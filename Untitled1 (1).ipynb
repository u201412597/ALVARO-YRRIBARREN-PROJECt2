```python
# Instalar yfinance y matplotlib si no están instalados
!pip install yfinance matplotlib

# Importar las bibliotecas necesarias
import yfinance as yf
import matplotlib.pyplot as plt

# Crear el objeto Ticker para Tesla
tesla = yf.Ticker("TSLA")

# Descargar los datos históricos de Tesla
tesla_data = tesla.history(period="max")

# Función para graficar los datos
def make_graph(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Close'], label='Precio de Cierre', color='blue')
    plt.title(title, fontsize=16)
    plt.xlabel('Fecha', fontsize=14)
    plt.ylabel('Precio de Cierre (USD)', fontsize=14)
    plt.legend()
    plt.grid()
    plt.show()

# Llamar a la función para graficar los datos de Tesla
make_graph(tesla_data, "Gráfica del Precio de las Acciones de Tesla")
```
